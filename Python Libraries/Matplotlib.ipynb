Line Plot :-

Line plots are a fundamental type of data visualization used in various fields for a variety of purposes like Trend Analysis,

Comparative Analysis, Highlighting Correlations, Forecasting, Visualizing Function Data, Data Exploration.



```python
# importing matplotlib library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
country_population = [700,750,740,730,752,800,790]

# Creating a Line Plot
plt.plot(country_population)
plt.show()
```


Markers in Matplotlib are used to highlight individual data points on a plot.

'o': Circle
 
'.': Point

',': Pixel

'x': X

'+': Plus

's': Square

'd': Diamond

'v': Triangle Down

'^': Triangle Up

'<': Triangle Left


'>': Triangle Right

'p': Pentagon

'h': Hexagon

```python
x_axis = [100,200,250,300,340,400,420]
plt.plot(x_axis,country_population,marker='*',ms=18,mfc='black')
# ms is an alias for markersize, which sets the size of the markers in a plot.
# mfc is an alias for markerfacecolor, which sets the fill color of the markers in a plot.
plt.show()
```

```python
plt.plot(x_axis,country_population,marker='*',ms=18,mfc='black',ls='dotted',lw=4) 
# ls is an alias for linestyle, which defines the style of the lines in a plot.
# lw is an alias for linewidth, which sets the width of the lines in a plot.
plt.show()
```

Bar Plot :-

Bar plots are a versatile and widely used type of plot in data visualization for displaying categorical data with rectangular bars. 

Each bar's length or height is proportional to the values it represents. Bar plots can be vertical or horizontal and are useful 

for comparing different groups or tracking changes over time.

```python
country_population = [450,700,900,500,800]
country_name = ['Japan','Bangladesh','India','Nepal','China']

# Creating a bar plot.
plt.bar(country_name,country_population)
plt.xlabel("Country_Name") # xlabel is used to set the label for the x-axis of a plot.
plt.ylabel("( in millions )") # ylabel is used to set the label for the y-axis of a plot.
plt.show()
```

```python
# color parameter is used to specify the color of various plot elements such as lines, markers, bars, and text. 
# ec = edge color
plt.bar(country_name,country_population,color='red',ec='k')
plt.xticks(rotation=90) #xticks is used to get or set the locations and labels of the x-axis ticks in a plot.
# rotation is used to rotate the label and yticks can also be used similarly.
plt.savefig('brand_population') # to save image with the specific name provided
plt.show()
```

Histogram :-

A histogram is a graphical representation of the distribution of numerical data, providing insights into various aspects of the data.

Visualizing Data Distribution, Identifying Patterns, Detecting Outliers, Summarizing Large Datasets etc.

```python
age = np.random.randint(1,100,500)
bin = [0,18,30,50,70,90,100]
# the bins parameter is used to define how the data in the age variable is grouped and displayed in the histogram
plt.hist(age,bins=bin,ec='k') # Creating a Histogram
# ec = edge color
plt.show()
```

```python
# If the bins parameter is not passed in the plt.hist function, Matplotlib will use a default method to determine the number and size of the bins.
plt.hist(age,ec='k')
plt.show()
```

Scatter Plot :-


A scatter plot is a type of data visualization that displays values for typically two variables as points on a Cartesian coordinate system.

The Scatter Plot is also used to check the relation(Positive os Negative) in graph.

```python
var1 = [25,41,63,96,45]
var2 = [45,63,65,25,75]
# creating a scatter plot

plt.scatter(var1,var2)

plt.xlabel("Var 1")
plt.ylabel("Var 2")

plt.show()
```

Pie Plot:-


A pie plot, or pie chart, is a circular statistical graphic divided into slices to illustrate numerical proportions.

Showing Proportions, Simplifying Data Presentation, Highlighting Major Components, Visual Appeal etc.

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
# Creating a Pie plot
plt.pie(country_population,labels=country_name,autopct="%1.2f%%")
plt.show()
```

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
exlposion = [0,0,0,0,0.1]
# The autopct parameter in a pie chart is used to display the percentage values on the pie chart slices.
# The explode parameter in a pie chart is used to highlight or "explode" one or more slices from the rest of the pie
plt.pie(country_population,labels=country_name,autopct="%1.2f%%",explode=exlposion)
plt.show()
```

```python

# The wedgeprops parameter in Matplotlib's plt.pie function is used to customize the appearance of the wedges (slices) in a pie chart. 
# It allows you to specify various properties of the wedges, such as their edge color, width, and line style.

country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
exlposion = [0,0,0,0,0.1]
plt.pie(country_population,labels=country_name,autopct="%1.2f%%",explode=exlposion,wedgeprops={'ec':'black','lw':2})
plt.show()
```

Box Plot :-


A box plot, or box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five key summary statistics: 

the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

 It is used for identifying outliers.

```python
df = pd.read_csv("Used_Bikes.csv")
df.drop_duplicates(inplace=True)
ktm = df[df['brand']=="KTM"]
```

```python
# Creating a Box Plot
plt.boxplot(ktm['price'])
plt.show()
```
