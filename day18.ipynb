```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries Imported!")
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# Rel Plot

sns.relplot(x='tip',y='total_bill',data=df)
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',style='smoker')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='time',col='day')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,col='day',kind='line')
plt.show()
```

```python
sns.relplot(x='size',y='total_bill',data=df,col='day',kind='line')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,col='day',kind='line')
plt.show()
```

```python
# Catplot
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex') # Strip
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='swarm') #  Swarm
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='box') # Box
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='violin')
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='boxen') # boxen
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='bar')
plt.show()
```

```python
sns.catplot(x='smoker',data=df,hue='sex',kind='count') # Count only one parameter is passed
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='point')
plt.show()
```
